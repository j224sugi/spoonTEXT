```python
import subprocess
import os
import shutil

gitProject = "C:/Users/syuuj/jsoup"
makedir = "logData"

if os.path.isdir(makedir):
    shutil.rmtree(makedir)
os.makedirs(makedir)

cmd = "git log --pretty=format:%H --name-only -- *.java"
hash = subprocess.run(
    cmd.split(), cwd=gitProject, capture_output=True, check=True, text=True
)
list = hash.stdout.split("\n\n")
k = 0
for commit in list:
    commit = commit.split("\n")
    newList = [i for i in commit if "test" not in i and "info" not in i]
    for i in range(1, len(newList)):
        newList[i] = gitProject + "/" + newList[i]
    fileWrite = "\n".join(newList)
    if len(newList) >= 2:
        k = k + 1
        with open("logData/" + str(k) + ".txt", "w", encoding="utf-8") as f:
            f.write(fileWrite)
```

```python
import os


def serchFile(ROOT_PATH):
    with open("first.txt", "w") as o:
        for pathname, dirnames, filenames in os.walk(ROOT_PATH):
            for filename in filenames:
                if filename.endswith(".java"):
                    fullPath = os.path.join(pathname, filename)
                    if (
                        "test" not in fullPath
                        and "info" not in fullPath
                        and "Exception" not in fullPath
                    ):
                        print(fullPath, file=o)


ROOT_PATH = "C:\\Users\\syuuj\\jsoup"

serchFile(ROOT_PATH)
```

git log : logを入手する
git log --stat : 変更されたファイルの一覧
git log --oneline : logを一行で表示する
git log --name-status : 変更したファイルを表示
git log --grep="fix" : fixが含まれるコミットログを取得
git rev-lst HEAD : Headのハッシュ値を全取得する
git rev-list --count HEAD : 現在ブランチのコミット数を得る
git show --name-status hash値　-- "*java" : 変更のあったjavaファイルを得る
git show hash値 pretty=format:"%H" --name-only →　変更のあったファイルとコミットIDを得る
git log --pretty=format:"%H" --name-only -- "*.java" : ハッシュ値と.javaで終わるファイル一覧を得る
